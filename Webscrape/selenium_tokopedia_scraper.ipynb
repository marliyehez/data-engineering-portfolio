```python
# Packages and Modules
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time 
import pandas as pd

# Website
website = "https://www.tokopedia.com/papercupdansedotan?source=universe&st=product&gclid=Cj0KCQiA14WdBhD8ARIsANao07hQfU-B4ywdPE6AjGXtdKYN5zxXz2R-w2B2iikK3EkKUHWQeJh0NJ4aArdeEALw_wcB"
```

```python
# Initialize Chrome
driver = webdriver.Chrome(r"C:\Users\User\Downloads\chromedriver_win32\chromedriver.exe")

# Open Website
driver.get(website)

# Maximize Window
driver.maximize_window()

# Scroll down to the bottom of the page
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
# Daftar nama dan harga produk
nama = driver.find_elements(By.XPATH, r'//div[@class="prd_link-product-name css-svipq6"]')
harga = driver.find_elements(By.XPATH, r'//div[@class="prd_link-product-price css-1ksb19c"]')
nama = [x.text for x in nama]
harga = [x.text for x in harga]
```

```python
# Ubah ke dalam bentuk dataframe
df = pd.DataFrame({'Nama Produk' : nama,
                    'Harga Produk' : harga})

df.tail()
```
